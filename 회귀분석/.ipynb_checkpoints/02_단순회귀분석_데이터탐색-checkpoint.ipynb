```python
import pandas as pd
```

```python
# header = 0 : 제목의 유무를 확인
df = pd.read_csv('./files_new/auto-mpg_v1.csv', header = 0)
```

```python
# 기초통계 : describe() = 통계 요약 조회
# 기초통계는 숫자만 가능 object는 불가능
# https://yganalyst.github.io/data_handling/Pd_5/
df.describe()
```

```python
# 데이터의 고유값 확인하기
df['horsepower'].unique()
```

```python
# 이상치 -> 결측치 -> 결측치 제거
# 결측치 : NaN을 의미 -> 결측치 제거 함수를 이용
# 이상치 : NaN이 아님 -> 결측치로 바꾸고 -> 결측치 제거 함수로 제거
```

```python
# '?'가 아닌 값인 행들만 가져와서 df[]에 넣기..!
df[df['horsepower'] != '?']
```

```python
df['horsepower'][df['horsepower'].isin(['?'])]
```

```python
import numpy as np
```

```python
# https://velog.io/@jen/Python-Pandas-dataframe%EC%9D%98-%ED%8A%B9%EC%A0%95-%EA%B0%92%EC%9D%84-NaN%EC%9C%BC%EB%A1%9C-%EB%B3%80%EA%B2%BD
# df의 0을 NaN 값으로 변경
df = df.replace('?', np.NaN)
```

```python
df['horsepower'].unique()
```

```python
df = df.dropna(axis = 0)
```

```python
df
```

```python
# inplace = True : 실제 메모리에 반영을 해라 = 삭제 후 메모리에 바로 반영하기
#  고로 df 데이터프레임에는 '?'가 사라진다.
df.dropna(subset = ['horsepower'], axis = 0, inplace = True)
```

```python
df
```

```python
df.info()
```

```python
# https://m.blog.naver.com/PostView.naver?isHttpsRedirect=true&blogId=youji4ever&logNo=221652121166
# object를 float 타입으로 변경
df = df.astype({'horsepower':'float'})
# df = df['horsepower'].astype('float')
```

```python
df.info()
```

```python
# 파일 저장하기
df.to_csv('./files_new/auto-mpg_v2.csv', index = False)
```

```python
# 회귀분석 = 독립변수, 종속변수이 있어야 함.
# 종속변수가 았다는 가정하..
# 단일, 다항, 다중 회귀분석 = 회귀분석은 3가지로 이루어져 있다
# 단일회귀분석(= 선형회귀분석) : (산점도의 모습이) 직선 / 하나의 독립변수를 가지고 종속변수에 영향이 있는지 확인
# 다중회귀분석 : 곡선 / 
# 다항회귀분석 : 상관관계를 확인할 때 주로 사용함. 
# 유의미하다 = 연관이 있다. = 상관관계상수(결정계수)가 1에 가까울수록 영향을 미친다.
# 음/양의 상관관계 = 그림의 형태 -> 관계 영향도
# 상관관계분석 -> 예측 => 지도/비지도 학습 -> 머신러닝(기계학습)(주로 지도학습) -> 딥러닝(주로 비지도학습)
# 예측모델 = 확률게임
# 예측 = 향후, 미래 => '머신러닝'이라는 지도학습의 모델을 이용해서 예측이 가능함. 
# 머신러닝은 한번에 하는 것이 아니라 여러 모델(함수)를 사용해서 결정계수가 가장 높은 모델을 적용해서 사용
```
