```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./files_new/auto-mpg_v2.csv')
```

```python
df
```

```python
# 지도학습 모델 : 단순회귀분석(선형회귀모델, 선형회귀예측모델)
# 종속변수가 항상 존재
# 직선의 방정식을 사용 : y = ax + b
# y : 종속변수 = 정답을 가지고 있는 것 = 정답
# x : 독립변수
# x 값으로 y값을 유추하는 것이라고 생각하면 된다. 
# a : 기울기
# b : y절편

# 학습 : fit = 학습모델 = 함수 = 알아서 계산해주는 것

# 공식은 정해져 있는데 a, b를 구하기 위해서 역산하는 것을 fit 이라고 생각하면 된다.
```

```python
# 사용할 변수들 : mpg, cylinders, horsepower, weight
# mpg(연비) : 종속변수 = 정답
# cylinders(실린더), horespower(엔진출력), weight(중량) : 독립변수
ndf = df[['mpg', 'cylinders', 'horsepower', 'weight']]
```

```python
ndf
```

```python
# 산점도 그래츠 그리기
# 변수 : mpg(종속변수), weight
```

```python
# 글씨체 변경.
# 한글깨짐 방지
plt.rc('font', family = 'Malgun Gothic')
# 타이틀 설정
plt.title('연비에 따른 중량')
# x축 라벨 설정
plt.xlabel('중량')
# y축 라벨 설정
plt.ylabel('연비')
# 산점도 그래프 그리기
plt.scatter(ndf.weight, ndf.mpg, color='lightpink')
```

```python
# 글씨체 변경.
# 한글깨짐 방지
plt.rc('font', family = 'Malgun Gothic')
# 타이틀 설정
plt.title('중량에 따른 연비')
# x축 라벨 설정
plt.xlabel('연비')
# y축 라벨 설정
plt.ylabel('중량')

# c : 색상 (파랑색)
# lw : 선의 굵기 (5 사이즈)
# ls : 선 스타일 (도트형)
# marker : 값의 위치에 표시할 마커 형태 (원형)
# ms : 마커의 전체 크기
# mec : 마커의 테두리 색상(녹색)
# mew : 마커 테두리 사이즈(5사이즈)
# mfc : 마커 내부의 채우기 색상(빨강)
# plt.plot([10, 20, 30, 40], [1, 4, 9, 16],
#          c="b", lw=5, ls="--", marker="o", ms=15, mec="g", mew=5, mfc="r")

# 산점도 그래프 그리기
plt.scatter(ndf.mpg, ndf.weight, color='lightpink')
```

```python
# 강사님
ndf.plot(kind = 'scatter',
         x = 'weight',
         y = 'mpg',
         c = 'red',
         s = 10,
         figsize = (10,5)
        )
plt.show()

# 그린 그래프를 메모리에서 제거하라
plt.close()
```

```python
# 해석
# 중량이 많이 나갈수록 연비는 떨어진다. = 음의 상관관계
# 중량은 연비에 영향을 미친다.
```

```python
# 회귀선(추세선) 그리기
ndf.plot(kind = 'scatter',
         x = 'weight',
         y = 'mpg',
         c = 'red',
         s = 10,
         figsize = (10,5)
        )

fit_weight = np.polyfit(ndf['weight'], df['mpg'], 1) # 'weight' 컬럼을 x값으로, 'mpg' 컬럼을 y값으로 하여 1차식으로 피팅한다.
trend_f = np.poly1d(fit_weight)

plt.plot(df['weight'], trend_f(df['weight']),"r-") # 회귀선(추세선)
# plt.title("y={:.6f}x+({:.6f})".format(fit_weight[0], fit_weight[1]))

plt.show()

# 그린 그래프를 메모리에서 제거하라
plt.close()
```

```python
# 강사님
# 회귀선(추세선)그리기
fig = plt.figure(figsize = (10, 5)) # 전체 사각형의 크기 지정

# 1번째 그래프 공간 만들기 (1행 2열의 1열에 넣어라)
ax1 = fig.add_subplot(1,2,1) 

# 2번째 그래프 공간 만들기 (1행 2열의 2열에 넣어라)
ax2 = fig.add_subplot(1,2,2)

# 1번째 그래프
sns.regplot(x = 'weight', 
            y = 'mpg',
            data = ndf,
            ax = ax1,
            color = 'red',
            marker = '+',
            fit_reg = True, # 회귀선(추세선)을 그려라
           )

# 2번째 그래프
sns.regplot(x = 'weight', 
            y = 'mpg',
            data = ndf,
            ax = ax2,
            color = 'blue',
            marker = 'x',
            fit_reg = False, # 회귀선(추세선)을 그리지 말아라
           )
```

```python
# jointplot 사용하여 회귀선 있는 것과 없는 것 그리기
# 한번에 안나오셔도 됩니다. 따로 그려도 된다..!
# jointplot() : 사람들이 가장 선호하는 그래프 형식이다!
# jointplot함수: 결합결과를 표시 
# kind="hex"속성: 표시 종류를 점이 아닌 6각형 색표시로 나타냄
sns.jointplot(x="weight", 
              y="mpg",
              color = 'green',
              data=ndf)

sns.jointplot(x="weight", 
              y="mpg",
              kind = 'reg',
              marker = '+',
              color = 'red',
              data=ndf)

sns.jointplot(x="weight", 
              y="mpg",
              data=ndf, 
              kind="hex")
```

```python
# 강사님
# 회귀선(추세선)그리기
fig = plt.figure(figsize = (10, 5)) # 전체 사각형의 크기 지정

# 1번째 그래프 공간 만들기 (1행 2열의 1열에 넣어라)
ax1 = fig.add_subplot(1,3,1) 

# 2번째 그래프 공간 만들기 (1행 2열의 2열에 넣어라)
ax2 = fig.add_subplot(1,3,2)

ax3 = fig.add_subplot(1,3,3)

# 1번째 그래프
sns.regplot(x = 'weight', 
            y = 'mpg',
            data = ndf,
            ax = ax1,
            color = 'red',
            marker = 'x',
            fit_reg = True, # 회귀선(추세선)을 그려라
           )

# 2번째 그래프
sns.regplot(x = 'cylinders', 
            y = 'mpg',
            data = ndf,
            ax = ax2,
            color = 'blue',
            marker = 'x',
            fit_reg = True,
           )

# 3번째 그래프
sns.regplot(x = 'horsepower', 
            y = 'mpg',
            data = ndf,
            ax = ax3,
            color = 'green',
            marker = 'x',
            fit_reg = True,
           )
```

```python
sns.jointplot(x="weight", 
              y="mpg",
              kind = 'reg',
              marker = '+',
              color = 'red',
              data=ndf)

sns.jointplot(x="cylinders", 
              y="mpg",
              kind = 'reg',
              marker = '+',
              color = 'blue',
              data=ndf)

sns.jointplot(x="horsepower", 
              y="mpg",
              kind = 'reg',
              marker = '+',
              color = 'green',
              data=ndf)
```

```python
# 컬럼과 컬럼 사이의 관계를 한번에 표현하는 그래프
# pairplot()
# https://pinkwink.kr/986
sns.pairplot(ndf,
             markers = '+',
             kind = 'reg'
            )

plt.show()

plt.close()

# 대각선은 1:1 관계를 의미
# 상관관계를 확인하기 위해서 사용함
```
