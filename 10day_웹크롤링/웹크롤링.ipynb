# 웹크롤링_멜론 사이트

```python
# 웹브라우저를 조작
# 웹브라우저에 보이는 HTML 태그속에서 텍스트 데기터를 분리해서 가지고오는 라이브러리
# selenium 설치 명령 : conda install -c conda-forge sleneium
# conda install로 설치가 안된다면 : pip install selenium
from selenium import webdriver
```

```python
# 구글 드라이버 다운로드 하기
# https://sites.google.com/a/chromium.org/chromedriver/downloads
# 크롬 버전 : 99.0.4844.51
```

```python
driver = webdriver.Chrome('C:/ChromeDriver_exe/chrome_99_driver.exe')
```

```python
# 웹 url로 접근하기
url = 'https://www.naver.com'
driver.get(url)
```

```python
# 웹페이지(html) 읽어 들이기
html = driver.page_source
html
```

### html 형태
### <body>와 <body> 사이의 내용 = 우리가 볼 수 있는 내용

#

    </head>

                </p>

    </body>
</html>
