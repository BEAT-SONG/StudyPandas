# 데이터 시각화 분석

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import font_manager, rc
import platform

plt.rc('font', family = 'Malgun Gothic')
```

```python
# 1. 기준월 및 기준일자별 버스 이용량 분석 비교
# 사용할 분석요소 : 승객구분, 기준월, 기준일
# 사용할 함수 : count() -> 빈도
df_bus_card_tot.head(1)
```

```python
# 피봇 데이터 -> 히트맵
# 피봇 데이터 <= 행에는 기준월, 컬럼에는 기준일, 데이터에는 이용량 = 승객구분에 대한 데이터를 빈도 체크한 것
```

```python
# 1.1 피봇 데이터 만들기
# 피봇 : 표
df_pivot = df_bus_card_tot.pivot_table(values  = '승객구분', # 데이터 값
                                       index   = '기준월', # 인덱스 / Y축
                                       columns = '기준일', # 컬럼 / X축
                                       aggfunc = "count") # 집계할 때 사용할 함수 / 합, 평균, 타운트 등..
                                       # , margins=True) 

df_pivot
```

```python
# 1.2 결측치 처리하기
df_pivot = df_pivot.fillna(0) # fillna() : 결측치를 채운다
df_pivot
```

```python
# 1.3 히트맵 그리기
# 그래프 크기 설정
plt.figure(figsize = (20,3)) # (너비, 높이)

#피봇 데이터 넣어서 그래프 그리기
sns.heatmap(df_pivot, 
            annot = True, # 데이터 값을 표현
            fmt = ".0f", # 실수형
            cmap = "rocket_r") # 히트맵의 형태(색상)

#타이틀 넣기
plt.title("기준월 및 기준일자별 버스 이용량 분석 비교")

#보여주기
plt.show()
```

```python
# 5. 시간대별 승객구분별 버스내체류시간(분) 분석
# 승객구분 = 범주형 데이터
```

```python
# 5.1 데이터 프레임 만들기
df_temp = pd.DataFrame()
df_temp['기준시간'] = df_bus_card_tot['기준시간']
df_temp['승객구분'] = df_bus_card_tot['승객구분']
df_temp['버스내체류시간'] = df_bus_card_tot['버스내체류시간(분)']

df_temp
```

```python
df_temp['승객구분'].unique()
```

```python
df_temp.drop_duplicates(['승객구분'])
```

```python
df_temp
```

```python
df_temp2 = (df_temp.groupby(['기준시간', '승객구분'], as_index = False).sum()).\
                sort_values(by = ['버스내체류시간'], ascending = False)
```

```python
df_temp2.head(10)
```

```python
# transpose() : 행과 열을 바꿔서 출력해주는 함수
df_temp2.head(10).transpose()
```

```python
# 5.2 막대 그래프 그리기
fig = plt.figure(figsize = (25,10))

sns.barplot(x = '기준시간',
            y = '버스내체류시간',
            hue = '승객구분',
            data = df_temp2)

plt.title('시간대별 승객구분별 버스내체류시간(분) 분석')

plt.show()
```

```python
# 6. 시간 및 승객구분별 빈도분석
```

```python
# 6.1 히스토그램 그래프 그리기
plt.figure(figsize = (12,4))
g = sns.histplot(data = df_temp2,
                 x = '기준시간',
                 bins = 30, # bin의 개수
                 kde = True, # 막대그래츠에 선 그리기(밀도 표현)
                 hue = '승객구분',
                 multiple = 'stack', # dodge
                 stat = 'density',# 비율로 표현하고자 할때 
                 shrink = 0.6)
plt.title('시간 및 승객구분별 빈도분석')
plt.xlabel('기준시간')
plt.ylabel('rate')
plt.show
```

```python
# 7. 승하차정류장별 버스내체류시간(분) 상위 30건 분석 비교
```

```python
df_bus_card_tot
```

```python
df_temp = pd.DataFrame()

df_temp['승하차정류장'] = df_bus_card_tot['승차정류장'] + '-->' + df_bus_card_tot['하차정류장']
df_temp['승객구분'] = df_bus_card_tot['승객구분']
df_temp['버스내체류시간'] = df_bus_card_tot['버스내체류시간(분)']

df_temp
```

```python
# 승하차정류장 및 승객구분별로 버스내체류시간의 sum을 구하기
df_temp_sel = (df_temp.groupby(['승하차정류장', '승객구분'], 
                            as_index = False).sum()).\
                            sort_values(by = ['버스내체류시간'], 
                                        ascending = False).head(30)
```

```python
df_temp_sel
```

```python
# 그래프 사이즈 지정
plt.figure(figsize = (12,4))

# 데이터 x,y 축 넣기
plt.plot(df_temp_sel['승하차정류장'], df_temp_sel['버스내체류시간'])

# 그래프 제목
plt.title('승하차정류장별 버스내체류시간(분) 상위 30건')
# x축 제목
plt.xlabel('승하차정류장')
# x축 표시값을 90도 수직으로 표현
plt.xticks(rotation = 90)
# y축 제목
plt.ylabel('버스내체류시간(분)')
# 그래프 내에 그리드선 표시 / 그리드 = 격자
plt.grid(True)

# 그래프 출력
plt.show()
```
