```python
# 년도와 월만 조회
df['기준년월'].str.split("-")
```

```python
df['기준년월'].str.extract('2010-01')
```

```python
# 강사님
df['년도'] = df['기준년월'].str.slice(0,4) # 년
df['월'] = df['기준년월'].str.slice(5,7) # 월
```

```python
df.head()
```

```python
df_filter = df[(df['국적'] == '홍콩')]

df_filter
```

```python
# 히트맵을 그리는 순서
# 1. 피봇으로 데이터 집계
# 2. 피봇 결과를 히트맵에 적용
```

```python
# 1. 피봇 데이터 만들기
# 피봇 : 표
df_pivot = df_filter.pivot_table(values = '관광', # 데이터 값
                      index = '년도', # 인덱스 / Y축
                      columns = '월') # 컬럼 / X축

df_pivot
```

```python
# 2. 히트맵 그리기
# 히트맵은 시본 라이브러리를 함께 사용해야 합니다,
import seaborn as sns
```

```python
# 그래프 크기 설정
plt.figure(figsize = (16, 10)) #정사각형에 가까운형태

#피봇 데이터 넣어서 그래프 그리기
sns.heatmap(df_pivot, annot = True, fmt = ".0f", cmap = "rocket_r")

#타이틀 넣기
plt.title("년/월별 [중국] 관광객 히트맵 시각화 분석")

#보여주기
plt.show()
```

```python
# 기준 만들기
cntry_list = ["중국", "일본", "대만", "미국", "홍콩"]

for cntry in cntry_list :
    
    condition = (df["국적"] == cntry)
    condition

    df_filter = df[condition]
    
    # 1. 피봇 데이터 만들기
    # 피봇 : 표
    df_pivot = df_filter.pivot_table(values = '관광', # 데이터 값
                      index = '년도', # 인덱스 / Y축
                      columns = '월') # 컬럼 / X축

    df_pivot
    
    # 그래프 크기 설정
    plt.figure(figsize = (16, 10)) #정사각형에 가까운형태

    #피봇 데이터 넣어서 그래프 그리기
    # annot = True : 값을 출력함 / annot = False : 값을 출력 안함
    # cmap : 색상을 지정
    sns.heatmap(df_pivot, annot = True, fmt = ".0f", cmap = "rocket_r") 

    #타이틀 넣기
    plt.title("년/월별 [{}] 관광객 히트맵 시각화 분석".format(cntry))

    #보여주기
    plt.show()
```
