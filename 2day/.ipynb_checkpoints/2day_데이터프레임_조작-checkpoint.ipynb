## 데이터프레임 행 조회 방법 2가지

numpy 라이브러리에서 loc = 인덱스(번호) 기준으로 행 데이터 읽기 / iloc = 행 번호를 기준으로 행 데이터 읽기

```python
# 데이터(전체) 조회
tips
```

### 2. numpy의 loc와 iloc 사용하기

```python
# 차이점을 확인하기 위해 특정 행 하나 삭제
tips_del = tips.drop([2])

tips_del
```

```python
# 행의 위치(순서)로 체크합니다.
tips_del.iloc[2]
```

```python
# 실제 인덱스(index) 번호로 체크합니다.
tips_del.loc[3]
```

```python
# 여러 데이터를 2차원으로 출력하기 위해서는 [[]] 대괄호가 2개 중첩되어야 한다.
tips.loc[[1, 3, 5]]
```

```python
tips.iloc[-1]
```

```python
# 위치값 : n-1값 까지의 데이터
tips[0:3]
```

```python
tips.iloc[0:3]
```

```python
tips.loc[0:3]
```

```python
# [행, [열]]
tips.loc[0:3, ["sex", 'day']]
```

## 정렬하기

```python
# axis[아식스] = 0(행) or 1(열)
# 행단위 정렬 : axis = 0
# 열단위 정렬 : axis = 1
# 기본적으로 ascending = True 오름차순, 안 적으면 기본이 오름차순
# ascending = False 내림차순
tips.sort_values(by = ['total_bill'], axis = 0, ascending = True)
```

```python
tips.sort_values(by = ['total_bill'], axis = 0, ascending = False)
```

```python
# 컬럼을 기준으로 정렬하기
tips.sort_index(axis = 1, ascending = True)
```

```python
tips.info()
```

```python
# 행단위로 전체 컬럼의 값들을 합하기
# 문자가 있는 것을 빼고 다 더함. = 빨간 경고문이 나옴
tips.sum(axis = 1)
```

```python
# 컬럼(열)단위로 전체 컬럼의 값들을 합하기
tips.sum(axis = 0)
```

## 데이터 (병)합하기

### 행단위로 합치기

```python
# 시리즈 데이터 합하기
# 시리즈 변수명 s1 ==> 값은 0, 1 / 인덱스 값은 A, B로 시리즈 만들기
# 시리즈 변수명 s2 ==> 값은 2, 3, 4 / 인덱스 값은 a, b, c로 시리즈 만들기
s1 = pd.Series([0, 1],
              index = ['A', 'B'])

s2 = pd.Series([2, 3, 4],
              index = ['a', 'b', 'c'])
```

```python
s1
```

```python
s2
```

```python
pd.concat([s1, s2], axis = 0)
```

### 컬럼 단위로 합치기

```python
df_temp1 = pd.DataFrame([['Dog', 3], ['Cat', 10], ['Tiger', 6], ['Monkey', 3]],
                       index = [0, 1, 2, 3],
                       columns = ['동물', '나이'])
```

```python
df_temp1
```

```python
df_temp2 = pd.DataFrame([['집', 0], ['초원', 1], ['수풀', 0], ['초원', 1]],
                       index = [0, 1, 2, 3],
                       columns = ['사는곳', '뿔의 갯수'])
```

```python
df_temp2
```

```python
pd.concat([df_temp1, df_temp2], axis = 1)
```

```python
df_info1 = pd.DataFrame({'고객번호' : [1001, 1002, 1003, 1004, 1005, 1006, 1007],
                        '이름' : ['강', '김', '홍', '이', '박', '최', '소']},
                       columns = ['고객번호', '이름'])

df_info1
```

```python
df_info2 = pd.DataFrame({'고객번호' : [1001, 1001, 1005, 1006, 1008, 1001],
                        '금액' : [10000, 20000, 15000, 5000, 100000, 30000]},
                       columns = ['고객번호', '금액'])

df_info2
```

```python
# 컬럼 단위로 옆으로 추가하기
# 단, 고객번호가 같은 값들만..
pd.merge(df_info1, df_info2, how = 'inner')
```

```python
pd.merge(df_info1, df_info2, how = 'outer')
```

```python
pd.merge(df_info1, df_info2, how = 'left')
```

```python
pd.merge(df_info1, df_info2, how = 'right')
```
